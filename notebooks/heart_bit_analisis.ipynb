# Medidas estadisticas

## Media

```python
import statistics
arr_media = []

for i in arrith_l:
  media = statistics.mean(i)
  arr_media.append(media)
```

##Kurtosis

```python
from scipy.stats import kurtosis

arr_kurtosis = []

for i in arrith_l:
  kurtosis_v = kurtosis(i)
  arr_kurtosis.append(kurtosis_v)
```
